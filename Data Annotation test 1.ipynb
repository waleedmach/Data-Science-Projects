```python
import requests
from bs4 import BeautifulSoup
import numpy as np

# Function to fetch and parse the publicly available Google Doc table
def fetch_google_doc_table(url):
    # Fetch the HTML content of the document
    response = requests.get(url)
    response.raise_for_status()  # Ensure the request was successful
    html_content = response.text

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table in the document (assuming the first table)
    table = soup.find('table')

    # Extract rows from the table
    table_data = []
    rows = table.find_all('tr')
    for row in rows:
        row_data = [cell.get_text().strip() for cell in row.find_all(['th', 'td'])]
        table_data.append(row_data)

    return table_data

# Function to build and print the grid based on the coordinates and characters
def print_character_grid(url):
    table_data = fetch_google_doc_table(url)
    
    # Extracting headers and ensuring proper column order
    headers = table_data[0]
    header_map = {header.lower(): idx for idx, header in enumerate(headers)}
    
    x_idx = header_map['x-coordinate']
    y_idx = header_map['y-coordinate']
    char_idx = header_map['character']
    
    # Extract the coordinate and character data
    coordinates = []
    for row in table_data[1:]:
        x = int(row[x_idx])
        y = int(row[y_idx])
        char = row[char_idx]
        coordinates.append((x, y, char))
    
    # Determine the size of the grid
    max_x = max(coord[0] for coord in coordinates)
    max_y = max(coord[1] for coord in coordinates)
    
    # Initialize the grid with spaces
    grid = np.full((max_y + 1, max_x + 1), ' ')
    
    # Place characters in the grid based on coordinates
    for x, y, char in coordinates:
        grid[y][x] = char
    
    # Reverse the rows to fix the upside-down issue
    grid = grid[::-1]
    
    # Print the grid
    for row in grid:
        print("".join(row))
```
